```python
import easyocr
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
reader = easyocr.Reader(['en'])
```

```python
# Load and display the image
image_path = 'img/45.jpg'
image = Image.open(image_path)
plt.imshow(image)
plt.axis('off')
plt.show()

# Perform OCR
result = reader.readtext(image_path)

# Create a plot with bounding boxes
fig, ax = plt.subplots(1, figsize=(12, 12))
ax.imshow(image)
plt.show()

if not result:
    print('null')
else:    
    for (bbox, text, prob) in result:
        print(f'{text}: {prob}')
    # Add bounding boxes
    for (bbox, text, prob) in result:
        # bbox contains coordinates of the bounding box
        (top_left, top_right, bottom_right, bottom_left) = bbox
        box_width = top_right[0] - top_left[0]
        box_height = bottom_left[1] - top_left[1]

        # Create a rectangle patch
        rect = patches.Rectangle((top_left[0], top_left[1]), box_width, box_height, linewidth=2, edgecolor='r', facecolor='none')
        ax.add_patch(rect)

        # Add text annotation
        plt.text(top_left[0], top_left[1] - 10, text, fontsize=15, color='red', bbox=dict(facecolor='white', alpha=0.8))

```
